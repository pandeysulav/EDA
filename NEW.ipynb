```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dog_breeds_traits.csv')
print(f"Number of rows: {df.shape[0]}, Number of columns: {df.shape[1]}")
print("\nColumn names:", df.columns.tolist())
```

```python
print("\nFirst 5 rows:\n", df.head())
print("\nData types and missing values:\n")
df.info()
print("\nSummary statistics:\n", df.describe())
```

```python
#This code helps to find out the unique features of every column.
print("\nUnique values in categorical columns:")
for col in ['size', 'origin_country', 'breed_group']:
   print(f"{col}:{df[col].nunique()} unique values")
   print(df[col].value_counts())
```

```python
#Use Pandasâ€™ nunique() across all columns to summarize diversity:
print("\nUnique values per column:\n", df.nunique())
```

```python
#Code to check if there are any missing values:
print("\n Missing values: \n", df.isnull().sum())
```
