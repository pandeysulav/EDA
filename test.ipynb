```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example data structure (replace with your actual data)
df = pd.DataFrame({
    "League Name": ["Premier League", "La Liga", "Bundesliga", "Serie A", "Ligue 1"],
    "Total Market Value": [900, 800, 750, 700, 650]
})
# df = pd.read_csv("your_file.csv")

# Sort data in descending order
df_sorted = df.sort_values("Total Market Value", ascending=False)

# Set figure size
plt.figure(figsize=(15, 12))

# Create horizontal barplot
sns.barplot(
    x="Total Market Value", 
    y="League Name", 
    data=df_sorted, 
    palette="viridis"
)

# Add value labels on bars
for index, value in enumerate(df_sorted["Total Market Value"]):
    plt.text(value + 10, index, f"{value:.2f}M", va='center')

# Set labels and title
plt.xlabel("Total Market Value (Million Euros)")
plt.ylabel("League Name")
plt.title("Total Market Value of Players in Each League")

# Display the plot
plt.tight_layout()
plt.show()
```
